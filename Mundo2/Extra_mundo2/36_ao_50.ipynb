```python
# 44
print("===== LIPS STORES =====")
try:
  preco = float(input('Digite o preço do produto R$: '))
except ValueError:
  print('Só aceitamos valores númericos')
else:
  print("""
  FORMAS DE PAGAMENTO
  [1] à vista dinheiro/cheque
  [2] à vista no cartão
  [3] 2x no cartão
  [4] 3x ou mais no cartão
  """)
  try:
    opcao = int(input('Digite sua opção (número): '))
  except ValueError:
    print('Só aceitamos valores númericos')
  else:
    if opcao == 1:
      desconto_10 = preco - (preco * 10 / 100)
      print(f'O produto custa {preco}R$ com 10% de desconto fica {desconto_10}R$')
    elif opcao == 2:
      desconto_5 = preco - (preco * 5 / 100)
      print(f'O produto custa {preco}R$ com 5% de desconto fica {desconto_5}R$')
    elif opcao == 3:
      parcelas = preco / 2
      print(f'O produto custa {preco}R$ com 2 parcelas fica {parcelas}R$')
    elif opcao == 4:
      juros = preco + (preco * 20 / 100)
      try:
        parcelas = int(input('Você vai pagar em quantas parcelas? '))
        preco_total = juros / parcelas
        print(f'O produto custa {preco}R$ com 20% de juros\
        \ndividido em {parcelas} parcelas fica {preco_total}R$')  
      except ValueError:
        print('Só aceitamos valores númericos')
    else:
      print('Opção Inexistente')
```

```python
# 45
from random import choice

maquina = ['pedra', 'papel', 'tesoura']
machine_jogo = choice(maquina)


print('''
Suas opções:

- Pedra
- Papel
- Tesoura
''')
try:
  jogador = input('Digite sua jogada: ').strip().lower()
except Exception as error:
  print('Error')
else:
  print('-=-=-=-'* 10)
  print('Computador jogou:', machine_jogo)
  print('Jogador jogou:', jogador)
  print('-=-=-=-'* 10)

if jogador == machine_jogo:
  print('Empate')
elif jogador == 'pedra' and machine_jogo == 'papel':
  print('Jogador Perdeu')
elif jogador == 'tesoura' and machine_jogo == 'pedra':
  print('Jogador Perdeu')
elif jogador == 'papel' and machine_jogo == 'tesoura':
  print('Jogador Perdeu')
else:
  print('Jogador Venceu')
```
