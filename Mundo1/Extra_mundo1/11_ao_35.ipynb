```python
# 28
from random import randint
from time import sleep


# Cabeçalho
print('-=-' * 20)
print('Vou pensar em um número entre 0 e 5 tente adivinhar')
print('-=-' * 20)

aleatorio = randint(1, 5)

# Entrada e Saída de dados
while True:
  try:
    jogador = int(input('Digite o número que estou pensando: '))
    print('Pensando...')
    sleep(2)
  except ValueError:
    print('Só aceito valores inteiros')
  else:
    if jogador < aleatorio:
      print('Penso um mais ;)')
    elif jogador > aleatorio:
      print('Penso um pouco menos ;)')
    elif jogador == aleatorio:
      print(f'Parabéns, acertou, estava pensando mesmo no {aleatorio} xD')
      break
```
