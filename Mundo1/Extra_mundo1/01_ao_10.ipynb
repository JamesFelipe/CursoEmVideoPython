```python
# 3 normal
n1 = int(input('Digite um valor inteiro: '))
n2 = int(input('Digite o mesmo/outro valor inteiro: '))
s = n1 + n2
print(f'{n1} + {n2} = {s}')
```

```python
# 3 diferente
try:
  n1 = int(input('Digite um valor inteiro: '))
  n2 = int(input('Digite o mesmo/outro valor inteiro: '))
  print(f'{n1} + {n2} = {n1 + n2}')
except ValueError:
  print('Aceito apenas valores inteiros')
```

```python
# 4
valor = input('Digite algo: ')
print(f'O tipo é: {type(valor)}')
print(f'Só tem espaços: {valor.isspace()}')
print(f'É um número? {valor.isnumeric()}')
print(f'É só alfabético? {valor.isalpha()}')
print(f'É só alfanúmerica? {valor.isalnum()}')
print(f'Está capitalizada? {valor.istitle()}')
print(f'Está tudo em minúsculo? {valor.islower()}')
print(f'Está tudo em maiúsculo? {valor.isupper()}')
```

```python
# 5
try:
  n = int(input('Digite um valor inteiro: '))
  print(f'o valor {n} tem o antecessor {n - 1} e o sucessor {n + 1}')
except ValueError:
  print('Aceitamos apenas valores inteiros')
```

```python
# 6
try:
  n = float(input('Digite um valor: '))
  print(f'O dobro de {n} é {n*2}')
  print(f'O triplo de {n} é {n*3}')
  print(f'a raiz quadrada de {n} é {(n**(1/2)):.2f}')
except ValueError:
  print('Só aceitamos números')
```

```python
# 6 com a biblioteca math
from math import sqrt
try:
  n = float(input('Digite um valor: '))
  print(f'O dobro de {n} é {n*2}')
  print(f'O triplo de {n} é {n*3}')
  print(f'a raiz quadrada de {n} é {sqrt(n):.2f}')
except ValueError:
  print('Só aceitamos números')
```

```python
# 7 
try:
  nota1 = float(input('Digite a primeira nota: '))
  nota2 = float(input('Digite a segunda nota: '))
  media = (nota1 + nota2) / 2
  print(f'Sua média final é: {media}')
except:
  print('Só aceitamos números')
```

```python
# 8
try:
  metros = float(input('Digite um valor em metros: '))
  print(f'Metros em centimeros: {metros * 100}')
  print(f'Metros em milimetros: {metros * 1000}')
except ValueError:
  print('Digite um número')
```

```python
# 9
try:
  n = int(input('Digite um número: '))
  print('-'* 15)
  print(f'{n} X 1 = {n * 1:>}')
  print(f'{n} X 2 = {n * 2:>}')
  print(f'{n} X 3 = {n * 3:>}')
  print(f'{n} X 4 = {n * 4:>}')
  print(f'{n} X 5 = {n * 5:>}')
  print(f'{n} X 6 = {n * 6:>}')
  print(f'{n} X 7 = {n * 7:>}')
  print(f'{n} X 8 = {n * 8:>}')
  print(f'{n} X 9 = {n * 9:>}')
  print(f'{n} X 10 = {n * 10:>}')
  print('-'* 15)
except:
  print('Digite um número inteiro')
```
